```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data for three experiments
operations = ['Op1', 'Op2', 'Op3']
experiments = ['Experiment1', 'Experiment2', 'Experiment3']

# Create dataframes for each experiment
df_exp1 = pd.DataFrame({'operations': operations})
df_exp1['Before'] = [100, 0, 300]
df_exp1['After'] = [0, 250, 350]
df_exp1['Optimized'] = [80, 180, 0]
df_exp1['Final'] = [75, 175, 275]

df_exp2 = pd.DataFrame({'operations': operations})
df_exp2['Before'] = [150, 200, 0]
df_exp2['After'] = [0, 300, 400]
df_exp2['Optimized'] = [90, 220, 0]
df_exp2['Final'] = [85, 280, 380]

df_exp3 = pd.DataFrame({'operations': operations})
df_exp3['Before'] = [200, 0, 400]
df_exp3['After'] = [0, 350, 450]
df_exp3['Optimized'] = [100, 260, 0]
df_exp3['Final'] = [95, 320, 420]

# Create figure with three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

# Define custom colors
custom_colors = {
    'Before': '#FF6B6B',    # Coral red
    'After': '#4ECDC4',     # Turquoise
    'Optimized': '#45B7D1', # Sky blue
    'Final': '#96CEB4'      # Sage green
}

# Function to create a single subplot
def create_subplot(data, ax, title):
    # Melt the dataframe
    df_melted = data.melt(id_vars='operations', 
                         value_vars=['Before', 'After', 'Optimized', 'Final'],
                         var_name='method',
                         value_name='Executiontime')
    
    # Create bar plot
    sns.barplot(data=df_melted, 
                y='operations', 
                x='Executiontime', 
                hue='method', 
                palette=custom_colors,
                ax=ax)
    
    # Get colors from plot
    colors = [p.get_facecolor() for p in ax.patches]
    
    # Add value labels
    for container, color in zip(ax.containers, colors):
        labels = [f'{v:.2f}' if v != 0 else '' for v in container.datavalues]
        ax.bar_label(container, 
                    labels=labels, 
                    color=color, 
                    padding=8, 
                    fontsize=10)
    
    # Customize subplot
    ax.grid(axis='x')
    ax.tick_params(axis='y', labelsize=12)
    ax.set_title(title)
    ax.set_xlabel('Execution Time (ms)')
    ax.set_ylabel('Operations')
    sns.despine()

# Create subplots for each experiment
create_subplot(df_exp1, ax1, 'Experiment 1')
create_subplot(df_exp2, ax2, 'Experiment 2')
create_subplot(df_exp3, ax3, 'Experiment 3')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
df_melted = df_test.melt(
    id_vars=['MissionId'],
    value_vars=None,
    var_name='Operation',
    value_name='Execution Time' 
)
print(df_melted)
melted_df = df_melted
print(melted_df)
```

```python
# Create figure with adjusted spacing
fig, ax = plt.subplots(figsize=(17, 15))  # Increased height for better spacing

# Create bar plot with adjusted parameters
sns.barplot(
    data=melted_df,
    y='Operation',
    x='Execution Time',
    hue='MissionId',
    palette='turbo',
    ax=ax,
    width=0.8,  # Reduced width for better spacing
    dodge=True   # Enable proper spacing between bars
)

# Get the colors used in the plot
colors = [p.get_facecolor() for p in ax.patches]

# Add value labels with matching colors
for container, color in zip(ax.containers, colors):
    # Create labels array, empty string for zeros
    labels = [f'{v:.2f}ms' if v != 0 else '' for v in container.datavalues]
    ax.bar_label(
        container,
        labels=labels,
        color=color,
        padding=8,
        fontsize=9,
        label_type='edge'  # Center labels on bars
    )

# Add grid with adjusted style
ax.grid(axis='x', linestyle='--', alpha=0.7)  # Dashed grid for better readability

# Customize appearance
ax.set_title('Operation Benchmark')
ax.set_xlabel('Execution Time (ms)', labelpad=50)
ax.set_ylabel('Operations')


# Adjust layout to prevent label cutoff
plt.tight_layout()

# Save and display
plt.savefig('mission_24450.jpg', bbox_inches='tight', dpi=300)
plt.show()
```
